```python
def groupingPLayers(dataframe):
  grouped_data = []
  current_group = []

  for index, row in dataframe.iterrows():
      current_group.append(row['Field1_text'])

      # Check if we've collected 11 elements
      if len(current_group) == 11:
          grouped_data.append({'Field1_text': current_group, 'Text': row['Text']})
          current_group = []

  grouped_df = pd.DataFrame(grouped_data)
  return grouped_df
```

```python
import pandas as pd

def groupingPLayers_missing_values(dataframe):
    grouped_data = []
    current_group = []

    prev_text = None
    for index, row in dataframe.iterrows():
        if prev_text is None or prev_text == row['Text']:
            current_group.append(row['Field1_text'])
        else:
            grouped_data.append({'Field1_text': current_group, 'Text': prev_text})
            current_group = [row['Field1_text']]
        prev_text = row['Text']

    if current_group:
        grouped_data.append({'Field1_text': current_group, 'Text': prev_text})

    grouped_df = pd.DataFrame(grouped_data)
    return grouped_df
```

```python
def combiner(dataframe_1,dataframe_2):
  concatenated_df = pd.concat([dataframe_1, dataframe_2], axis=1)
  return concatenated_df
```

# Calculating the performance index of each player

```python
players_df = final_players_performance_scores.copy(deep=True)

teams_df = sorted_data_frame.copy(deep=True)
teams_df.drop(["ID","MatchNumber","SuperOver","WonBy","Margin","Player_of_Match"],axis=1,inplace=True)
teams_df['Season'] = teams_df['Season'].astype(int)

def format_name(name):
    formatted_name = ' '.join([part.title() for part in name.split()])
    return formatted_name
def capitalize_names(name_list):
    temp=name_list.split(' ')
    temp2=[name.title() for name in temp]
    return ' '.join(temp2)

players_df['Player'] = players_df['Player'].apply(capitalize_names)

teams_df['players_team1'] = teams_df['players_team1'].apply(lambda x: [format_name(name) for name in x])

teams_df['players_team2'] = teams_df['players_team2'].apply(lambda x: [format_name(name) for name in x])


def get_scaled_score(player, year):
  total_score = 0.0
  total_weight = 0.0
  for index, row in players_df.iterrows():
    if row['Player'] == player and row['Year'] <=year:
      weight = 0.8

      if row['Year']==year:
        weight = 0.3
      if row['Year'] == year - 1:
        weight = 0.6
      elif row['Year'] < (year - 1):
        weight= 0.1


      total_score += row['Scaled_Score'] * weight
      total_weight += weight
  if total_score==0:
      return 0.1
  if total_weight > 0:
    return total_score / total_weight
  return None

for index, row in teams_df.iterrows():
  team1_scores = []
  for player in row['players_team1']:
      score = get_scaled_score(player, row['Season'])
      if score is not None:
          team1_scores.append(score)
  teams_df.at[index,'Total_Team1_Performance_score']=sum(team1_scores)

  team2_scores = []
  for player in row['players_team2']:
      score = get_scaled_score(player, row['Season'])
      if score is not None:
          team2_scores.append(score)
  teams_df.at[index,'Total_Team2_Performance_score']=sum(team2_scores)

  print(f"Team 1 Scores: {team1_scores}")
  print(f"Team 2 Scores: {team2_scores}")
  print("---------------")
```
